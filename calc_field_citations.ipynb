```python
import json
import os
from dotenv import dotenv_values
from collections import defaultdict
import csv
```

```python
# Environment variables in a .env file
# S2_API_KEY - For api access to dataset links
# DATA_DIR - directory to download data to (e.g. external harddrive)
config = dotenv_values()
```

### Load Fields of study

```python
field_dir = config['DATA_DIR'] + 'paper_field'
#Loop through 30 files
paper_field = {}
for filename in os.listdir(field_dir):
    print(filename)
    with open(field_dir + '/' + filename) as f:
        paper_field.update(json.load(f))
```

```python
len(paper_field)
```

### Process & count citations

```python
citations_dir = config['DATA_DIR'] + 'citations_tsv'
#Loop through 30 files
field_cites_field = defaultdict(lambda: defaultdict(int))
for filename in os.listdir(citations_dir):
    print(filename)
    with open(citations_dir + '/' +  filename) as f:
        for i, l in enumerate(f):
            line = l.strip('\n').split('\t')
            if(i % 100000 == 0):
                print(".", end="", flush=True)
            citing = line[0]
            cited = line[1]
            if(citing in paper_field and cited in paper_field):
                for citing_field in paper_field[citing]:
                    for cited_field in paper_field[cited]:
                        field_cites_field[citing_field][cited_field] +=1
        print('')
```

```python
with open(config['DATA_DIR'] + 'field_cites.json', 'w') as f_out:
    json.dump(field_cites_field, f_out)
```
